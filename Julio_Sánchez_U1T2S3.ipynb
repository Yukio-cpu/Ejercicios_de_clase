```python
#Definir una variable global llamada pais.
#Crear un procedimiento datos_personales donde se solicite:
#- Nombre, edad, genero , altura(m), peso(kg).
#Se requiere mostrar los datos ingresado y calcular el indice de masa corporal.
#IMC= PESO/ALTURA**2
#- Ademas mediante uso de estructuras de control clasique si el peso es correcto si:
#- IMC <19 -> Bajo peso
#- IMC esta entre 19 <IMC< 30 -> Peso Normal
#- IMC >30 Sobrepeso

pais = "Ecuador"
def datos_personales():
    nombre = input("Ingrese su nombre: ")
    edad = int(input("Ingrese su edad: "))
    genero = input("Ingrese su genero: ")
    altura = float(input("Ingrese su altura en metros: "))
    peso = float(input("Ingrese su peso en kg: "))

    imc = peso / (altura ** 2)

    print(f"Nombre: {nombre}")
    print(f"Edad: {edad}")
    print(f"Género: {genero}")
    print(f"Altura: {altura} m")
    print(f"Peso: {peso} kg")
    print(f"Índice de Masa Corporal (IMC): {imc:.2f}")

    if imc < 19:
        print("Bajo peso")
    elif 19 <= imc < 30:
        print("Peso Normal")
    else:
        print("Sobrepeso")
```

Pide el monto de una compra. 
Si la compra es mayor o igual a 50, aplica un descuento del 5%. 
Caso contrario, muestra el total sin descuento.

```python
#EJERCICIO #2
#Pide el monto de una compra. 
#Si la compra es mayor o igual a 50, aplica un descuento del 5%. 
#Caso contrario, muestra el total sin descuento.

costo = float(input("Ingrese el monto de la compra: "))

if costo >= 50:
    descuento = costo * 0.05
    total_con_descuento = costo - descuento
    print("Aplicas para descuento del 5%")
    print(f"Total con descuento: {total_con_descuento:.2f}")
else:
    print(f"Total sin descuento: {costo:.2f}")
```

```python
#Desarrolla un programa que permita ingresar el **nombre de un estudiante** y su **calificación final** sobre 10.
#El programa debe mostrar:
#- el nombre del estudiante,
#- la calificación ingresada,
#- y un mensaje según el resultado:
#| Rango | Mensaje |
#|---|---|
#| 9 a 10 | Excelente |
#| 7 a menor que 9 | Aprobado con buen desempeño |
#| 5 a menor que 7 | Aprobado con dificultad |
#| menor que 5 | Reprobado |
### Requisito adicional
#Si la nota ingresada es menor que 0 o mayor que 10, el programa debe mostrar:
#**“Calificación no válida”**

nombre_estudiante = input("Ingrese el nombre del estudiante: ")
calificacion = float(input("Ingrese la calificación final (sobre 10): "))
# Validar la calificación
if 0 <= calificacion <= 10:
    print(f"Nombre del estudiante: {nombre_estudiante}")
    print(f"Calificación ingresada: {calificacion:.2f}")
    if 9 <= calificacion <= 10:
        print("Excelente")
    elif 7 <= calificacion < 9:
        print("Aprobado con buen desempeño")
    elif 5 <= calificacion < 7:
        print("Aprobado con dificultad")
    else:
        print("Reprobado")
else:
    print("Calificación no válida")
```
