## Enunciado Ejercicio integrador: Control de compra en una tienda

Desarrolle un programa en Python que permita registrar los datos de una compra realizada por un cliente en una tienda.
El programa debe cumplir con lo siguiente:
Defina una variable global llamada tienda que almacene el nombre del negocio. Cree una función llamada registrar_compra() que solicite los siguientes datos: nombre del cliente edad nombre del producto cantidad de productos precio unitario Dentro de la función debe: mostrar los datos ingresados;
calcular el subtotal de la compra:
subtotal = cantidad * precio_unitario
aplicar un descuento según el subtotal usando estructuras de control: si el subtotal es menor a 20, no aplica descuento; si el subtotal está entre 20 y menor que 50, aplicar 5%; si el subtotal está entre 50 y menor o igual a 100, aplicar 10%; si el subtotal es mayor a 100, aplicar 15%. Calcular y mostrar: subtotal descuento total a pagar También evaluar la edad del cliente: si tiene 18 años o más, mostrar que es mayor de edad; caso contrario, mostrar que es menor de edad. Finalmente, mostrar un mensaje con el nombre de la tienda.

```python
tienda = "Mi Tienda"
#Definir la función para registrar la compra
def registrar_compra():
    nombre_cliente = input("Ingrese el nombre del cliente: ")
    edad_cliente = int(input("Ingrese la edad del cliente: "))
    nombre_producto = input("Ingrese el nombre del producto: ")
    cantidad_productos = int(input("Ingrese la cantidad de productos: "))
    precio_unitario = float(input("Ingrese el precio unitario del producto: "))

    #Calculo del subtotal
    subtotal = cantidad_productos * precio_unitario

    #Descuento según el subtotal
    if subtotal >= 20 and subtotal < 50:
        descuento = subtotal * 0.05
    elif subtotal >= 50 and subtotal <= 100:
        descuento = subtotal * 0.10
    elif subtotal > 100:
        descuento = subtotal * 0.15 
    else:
        descuento = 0

    #Calculo del total a pagar
    total_a_pagar = subtotal - descuento

    #Mostrar los datos ingresados y los cálculos
    print("***** Datos de la compra *****")
    print(f"Nombre del cliente: {nombre_cliente}")
    print(f"Edad del cliente: {edad_cliente}")
    print(f"Nombre del producto: {nombre_producto}")
    print(f"Cantidad de productos: {cantidad_productos}")
    print(f"Precio unitario: ${precio_unitario:.2f}")
    print(f"Nombre de la tienda: {tienda}")
    
    #Mostrar el valor del producto
    print("\n***** Precio del producto *****")
    print(f"Subtotal: ${subtotal:.2f}")
    print(f"Descuento: ${descuento:.2f}")
    print(f"Total a pagar: ${total_a_pagar:.2f}")

    #Verificar la edad del cliente
    if edad_cliente >= 18:
        print("El cliente es mayor de edad.")
    else:
        print("El cliente es menor de edad.")

registrar_compra()
```

Escriba un programa que defina un número secreto fijo (por ejemplo, 9), o lo puede ingresar por teclado. El usuario tiene hasta 5 intentos
para adivinar el número. Por cada intento: Se incrementa un contador de intentos. Se compara el número ingresado con el secreto: Si es
igual: se imprime "¡Correcto! Lo lograste en X intentos" y el programa termina. o Si no: Usa un operador ternario para mostrar: o
"Demasiado alto" si el número es mayor. o "Demasiado bajo" si es menor. o Si el usuario falla los 5 intentos, el programa muestra: "Lo
siento, se agotaron los intentos. El número era X". Resuelve

```python
import random
numero_secreto = random.randint(0, 10)
intentos = 0
max_intentos = 5
while intentos < max_intentos:
    numero = int(input("Adivina el número secreto: "))
    intentos += 1
    if numero == numero_secreto:
        print(f"¡Correcto! Lo lograste en {intentos} intentos")
        break
    else:
        print("Demasiado alto" if numero > numero_secreto else "Demasiado bajo")
if intentos >= max_intentos:
    print(f"Lo siento, se agotaron los intentos. El número era {numero_secreto}")
```
