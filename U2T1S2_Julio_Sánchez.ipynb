- DADA UNA FUNCION REALICE LA 4 OPERACIONES BASICAS PARA QUE TOME POR PARAMETROS X, Y.
- ADICIONAL SE ENVIAN ARGUMENTOS QUE TIENE QUE EL USUARIO INGRESAR POR TECLADO.

```python
def operaciones (x,y):
    suma = x + y
    resta = x - y
    multiplicacion = x * y
    division = x / y
    return suma, resta, multiplicacion, division

def promedio():
    suma,_,_,_ = operaciones(x,y)
    promedio = suma / 2
    if promedio > 9 :
        evaluacion = "Excelente"
    else:
        evaluacion = "Buena"
    return promedio, evaluacion

#Programa Principal
print("Ingrese los valores de X y Y")
x = int(input("Valor de X: "))
y = int(input("Valor de Y: "))
nombre = input("Ingrese el nombre del alumno: ")

x,y,z,p = operaciones(x,y) 
print("*****OPERACIONES*****")
print("Suma:", x)
print("Resta:", y)
print("Multiplicación:", z)
print("División:", p)
print("Promedio:", promedio()[0])
print("Evaluación:", promedio()[1])
print("el nombre del alumno es :", nombre)
```

Crear una funcion que calcule el precio final de un producto aplicando un descuento. Si no se indica descuento se aplica por defecto.

```python
def aplicar_descuento(precio, descuento=0.10):
    valor_descuento = precio * descuento
    precio_final = precio - valor_descuento
    return precio_final

p1 = float(input("Ingrese el precio del producto: "))
x = aplicar_descuento(p1)
print("Precio con descuento por defecto:", x)
print("Precio con descuento personalizado:", aplicar_descuento(x, 0.25))
```

```python
def calcular_promedio(*notas):
    suma = 0
    cont = 0
    for nota in notas:
        print("Nota:", nota)
        suma += nota
        cont += 1
    promedio = suma / cont
    return suma, promedio

print("La suma es :", calcular_promedio(8, 9, 10)[0])
print("El promedio es :", calcular_promedio(8, 9, 10)[1])
```

Desarrollar un ejercicio que permita ingresar n precios de los productos, y que calcule el valor a pagar total mas IVA=0,12 por defecto. Utilice la funcion *args y que los datos se ingresen por teclado.

```python
# Ejercicio 2
# Desarrollar un ejercicio que permita ingresar n precios de los productos, y que calcule el valor a pagar total mas IVA=0,12 por defecto.
# Utilice la funcion *args y que los datos se ingresen por teclado.

def calcular_total_con_iva(*precios, iva=0.12):
    """Calcula el total con IVA sobre los precios recibidos."""
    subtotal = sum(precios)
    total = subtotal * (1 + iva)
    return subtotal, total

n = int(input('¿Cuántos precios deseas ingresar? '))
precios = []
for i in range(1, n + 1):
    precio = float(input(f'Ingrese el precio del producto {i}: '))
    precios.append(precio)

subtotal, total = calcular_total_con_iva(*precios)
print(f'Precio total sin IVA: {subtotal:.2f}')
print(f'Precio total con IVA (12%): {total:.2f}')
```

```python
def procesar_dato(dato):
    if isinstance(dato, int):
        return f"Dato numerico. Su doble es: {dato * 2}"
    elif isinstance(dato, str):
        return f"Dato de texto. Texto en mayusculas: {dato.upper() }"
    elif isinstance(dato, list):
        return f"Dato tipo lista. Cantidad de elementos: {len(dato)}"
    else:
        return "Tipo de dato no reconocido"
    
print(procesar_dato(1.5)) # integer
print(procesar_dato("programación modular")) #str
print(procesar_dato([8, 9, 10])) #list
```
