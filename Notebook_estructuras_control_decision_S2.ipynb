## 1. Introducción

Las estructuras de control permiten que un programa tome decisiones dependiendo de una condición.  
En programación, esto ayuda a que el sistema responda de manera diferente según los datos ingresados por el usuario.


## 2. Definición

Las **estructuras de control de decisión** permiten ejecutar bloques de código dependiendo de si una condición es **verdadera** o **falsa**.


## 3. Tipos de estructuras de decisión

### 3.1 Estructura `if`
Ejecuta un bloque de instrucciones solo si la condición es verdadera.


```python
edad = 18

if edad >= 18:
    print('Eres mayor de edad')
```

### 3.2 Estructura `if - else`
Permite ejecutar una acción cuando la condición es verdadera y otra cuando es falsa.


```python
edad = 16

if edad >= 18:
    print('Eres mayor de edad')
else:
    print('Eres menor de edad')
```

### 3.3 Estructura `if - elif - else`
Sirve para evaluar varias condiciones.


```python
nota = 85

if nota >= 90:
    print('Excelente')
elif nota >= 70:
    print('Aprobado')
else:
    print('Reprobado')
```

## 4. Ejercicios prácticos guiados

### Ejercicio 1: Número positivo
Solicita un número e indica si es positivo.


```python
numero = int(input('Ingrese un número: '))

if numero > 0:
    print('El número es positivo')
elif numero == 0:
    print('El número es cero')
else:
    print('El número es negativo')
```

### Ejercicio 2: Número par o impar
Determina si un número es par o impar.


```python
numero = int(input('Ingrese un número: '))

if numero % 2 == 0:
    print('El número es par')
else:
    print('El número es impar')
```

### Ejercicio 3: Verificar si puede votar
Determina si una persona puede votar según su edad.


```python
edad = int(input('Ingrese su edad: '))

if edad >= 18:
    print('Puede votar')
else:
    print('No puede votar')
```

### Ejercicio 4: Clasificación de nota
Clasifica una nota en excelente, bueno, regular o reprobado.


```python
nota = float(input('Ingrese su nota: '))

if nota >= 90:
    print('Excelente')
elif nota >= 70:
    print('Bueno')
elif nota >= 50:
    print('Regular')
else:
    print('Reprobado')
```

## 5. Actividad práctica integradora

Desarrolle un programa que solicite el nombre del estudiante y tres calificaciones.  
Luego calcule el promedio y muestre si el estudiante aprueba o reprueba.


```python
nombre = input('Ingrese su nombre: ')
nota1 = float(input('Ingrese la nota 1: '))
nota2 = float(input('Ingrese la nota 2: '))
nota3 = float(input('Ingrese la nota 3: '))

promedio = (nota1 + nota2 + nota3) / 3

print(f'Estudiante: {nombre}')
print(f'Promedio: {promedio:.2f}')

if promedio >= 70:
    print('Estado: Aprobado')
else:
    print('Estado: Reprobado')
```
